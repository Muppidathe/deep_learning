```python
from keras.datasets import cifar10
(x_train,y_train),(x_test,y_test)=cifar10.load_data()
x_train=x_train/255.0
x_test=x_test/255.0
y_test
```

```python
from tensorflow.keras.utils import to_categorical # onehot encoding
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
y_test
```

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input,decode_predictions
```

```python
model=InceptionV3(include_top=False,input_shape=(224,224,3))
```

```python
for layer in model.layers:
  print(layer)
  layer.trainable=False
```

```python
for layer in model.layers:
  print(layer.name,' ',layer.trainable)
```

```python
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense,UpSampling2D,Resizing,Flatten,Dropout
new_model=Sequential()
new_model.add(UpSampling2D((2,2)))
new_model.add(UpSampling2D((2,2)))
new_model.add(UpSampling2D((2,2)))
new_model.add(Resizing(224,224,interpolation='bilinear'))
new_model.add(model)
new_model.add(Flatten())
new_model.add(Dense(128,activation='relu'))
new_model.add(Dropout(.2))
new_model.add(Dense(10,activation='softmax'))
```

```python
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.losses import categorical_crossentropy
new_model.compile(optimizer=RMSprop(learning_rate=.01),loss=categorical_crossentropy,metrics=['accuracy','recall'])
new_model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test),verbose=2)
```
