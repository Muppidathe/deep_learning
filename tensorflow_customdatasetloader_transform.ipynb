```python
import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory
data_dir='/content/kaggle-cat-vs-dog-dataset/kagglecatsanddogs_3367a/PetImages'
train_dataloader=image_dataset_from_directory(data_dir,batch_size=32,seed=42,validation_split=.2,subset='training',image_size=(128,128))
test_dataloader=image_dataset_from_directory(data_dir,batch_size=32,seed=42,validation_split=.2,subset='validation',image_size=(128,128))
```

```python
normalization_layer = tf.keras.layers.Rescaling(1./255)
test_dataloader_mapped=test_dataloader.map(lambda x,y:(normalization_layer(x),y))
```

```python
for image,label in test_dataloader_mapped:
  print(label.shape)
  print(image.shape)
```

```python
import matplotlib.pyplot as plt
for image,label in test_dataloader_mapped:
  print(test_dataloader.class_names[label[0]])
  print(image[0].shape)
  print(label[0])
  plt.imshow(image[0])

  break
```
